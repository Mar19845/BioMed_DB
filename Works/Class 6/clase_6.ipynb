```python
## Librerias
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib

```python
np.random.rand(20)
```

```python
plt.plot(np.random.rand(20))
plt.title('Titulo prueba')
plt.show()
```

## Partes del objeto Figure 
<img src="https://matplotlib.org/stable/_images/anatomy.png">

```python
fig = plt.figure()  # figure vacia sin Axes
```

```python
fig, ax = plt.subplots()  # una figure con un Axes simple
```

```python
x = np.linspace(-np.pi, np.pi, 101)
y = np.sin(x) + np.sin(3*x)/3.0

plt.plot(x, y) 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico simple')
plt.show()
```

```python
fig, axs = plt.subplots(2,2)  # una figure con un grupo de Axes de 2x2
```

```python
fig, axes = plt.subplots(2)
axes[0].plot(np.random.rand(20))
axes[1].plot(np.random.rand(200))
plt.show()
```

```python
plt.plot(np.random.rand(20), marker='o')
plt.show()
```

```python
plt.plot(np.random.rand(20), linestyle = ':')
plt.show()
```

```python
plt.plot(np.random.rand(20), marker='o')

plt.xlabel("eje x")
plt.ylabel("eje y")
plt.show()
```

```python
plt.plot(np.random.rand(20), marker='o')

plt.xlabel("eje x")
plt.ylabel("eje y")
plt.grid()
plt.show()
```

```python
x =np.random.rand(20)
y =np.random.rand(20)
plt.scatter(x, y)

plt.xlabel("eje x")
plt.ylabel("eje y")
plt.grid()
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.grid()
plt.show()
```

```python
etiquetas = np.array(["A", "B", "C", "D"])
x = np.array(np.random.randint(1,100,4))

plt.pie(x, labels=etiquetas)
plt.show() 
```
